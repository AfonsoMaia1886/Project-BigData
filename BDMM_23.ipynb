```python
MISSING_STRINGS = {"", "na", "n/a", "nan", "none", "null", "missing"}

def is_missing(x: Any) -> bool:
    if x is None:
        return True
    if isinstance(x, float) and math.isnan(x):
        return True
    if isinstance(x, str) and x.strip().lower() in MISSING_STRINGS:
        return True
    return False

def clean_string(x: Any) -> Optional[str]:
    if is_missing(x):
        return None
    s = str(x).strip()
    s = re.sub(r"\s+", " ", s)
    return s or None

def parse_decimal128(x: Any) -> Optional[Decimal128]:
    if is_missing(x):
        return None
    s = clean_string(x)
    if s is None:
        return None
    s = s.replace("$", "").replace("€", "").replace(",", "")
    try:
        return Decimal128(s)
    except (InvalidOperation, Exception):
        return None

def parse_int(x: Any) -> Optional[int]:
    if is_missing(x):
        return None
    if isinstance(x, bool):
        return int(x)
    try:
        return int(float(str(x).strip()))
    except Exception:
        return None

def parse_float(x: Any) -> Optional[float]:
    if is_missing(x):
        return None
    try:
        return float(str(x).strip())
    except Exception:
        return None

def parse_date(x: Any) -> Optional[datetime]:
    if is_missing(x):
        return None
    if isinstance(x, datetime):
        dt = x
    elif isinstance(x, pd.Timestamp):
        dt = x.to_pydatetime()
    else:
        try:
            dt = dtparser.parse(str(x))
        except Exception:
            return None

    if dt.tzinfo is None:
        dt = dt.replace(tzinfo=timezone.utc)
    else:
        dt = dt.astimezone(timezone.utc)
    return dt

def normalize_bool(x: Any) -> Optional[bool]:
    if is_missing(x):
        return None
    if isinstance(x, bool):
        return x
    s = str(x).strip().lower()
    truthy = {"t", "true", "yes", "y", "1"}
    falsy = {"f", "false", "no", "n", "0"}
    if s in truthy:
        return True
    if s in falsy:
        return False
    return None

def unique_clean_list(values: Any) -> list[str]:
    if not isinstance(values, list):
        return []
    seen = set()
    out = []
    for value in values:
        s = clean_string(value)
        if s is not None and s not in seen:
            seen.add(s)
            out.append(s)
    return out

def clean_coords(location_doc: Optional[dict]) -> Optional[dict]:
    if not isinstance(location_doc, dict):
        return None

    coords = location_doc.get("coordinates")
    if not isinstance(coords, list) or len(coords) != 2:
        return None

    lon = parse_float(coords[0])
    lat = parse_float(coords[1])

    if lon is None or lat is None:
        return None
    if not (-180 <= lon <= 180 and -90 <= lat <= 90):
        return None

    return {"type": "Point", "coordinates": [lon, lat]}

def safe_hash_text(text: Any) -> Optional[str]:
    s = clean_string(text)
    if s is None:
        return None
    return hashlib.sha256(s.encode("utf-8")).hexdigest()

def bounded_int(x: Any, low: int, high: int) -> Optional[int]:
    value = parse_int(x)
    if value is None or not (low <= value <= high):
        return None
    return value

def bounded_float(x: Any, low: float, high: float) -> Optional[float]:
    value = parse_float(x)
    if value is None or not (low <= value <= high):
        return None
    return value
```

## 5. Document reshaping functions

```python

DETAIL_FIELDS = [
    "description",
    "neighborhood_overview",
    "notes",
    "transit",
    "access",
    "interaction",
    "house_rules",
    "summary",
    "space",
]

REVIEW_SCORE_FIELDS = [
    "review_scores_accuracy",
    "review_scores_cleanliness",
    "review_scores_checkin",
    "review_scores_communication",
    "review_scores_location",
    "review_scores_value",
    "review_scores_rating",
]

def build_host_doc(doc: dict) -> Optional[dict]:
    host = doc.get("host", {})
    host_id = host.get("host_id")
    if is_missing(host_id):
        return None

    return {
        "_id": clean_string(host_id),
        "host_name": clean_string(host.get("host_name")),
        "host_since": parse_date(host.get("host_since")),
        "host_location": clean_string(host.get("host_location")),
        "host_response_time": clean_string(host.get("host_response_time")),
        "host_response_rate": clean_string(host.get("host_response_rate")),
        "host_is_superhost": normalize_bool(host.get("host_is_superhost")),
        "host_has_profile_pic": normalize_bool(host.get("host_has_profile_pic")),
        "host_identity_verified": normalize_bool(host.get("host_identity_verified")),
        "host_listings_count": parse_int(host.get("host_listings_count")),
        "host_total_listings_count": parse_int(host.get("host_total_listings_count")),
        "host_verifications": unique_clean_list(host.get("host_verifications")),
        "schema_version": 1,
    }

def build_host_snapshot(doc: dict) -> Optional[dict]:
    host = doc.get("host", {})
    host_id = clean_string(host.get("host_id"))
    if host_id is None:
        return None

    return {
        "host_id": host_id,
        "host_name": clean_string(host.get("host_name")),
        "host_since": parse_date(host.get("host_since")),
        "is_superhost": normalize_bool(host.get("host_is_superhost")),
        "response_rate": clean_string(host.get("host_response_rate")),
    }

def compute_review_stats(doc: dict) -> dict:
    reviews = doc.get("reviews", [])
    clean_review_dates = []
    for r in reviews if isinstance(reviews, list) else []:
        d = parse_date(r.get("date"))
        if d is not None:
            clean_review_dates.append(d)

    stats = {
        "count": len(reviews) if isinstance(reviews, list) else 0,
        "last_review_date": max(clean_review_dates) if clean_review_dates else None,
    }

def compute_review_stats(doc: dict) -> dict:
    reviews = doc.get("reviews", [])
    clean_review_dates = []
    for r in reviews if isinstance(reviews, list) else []:
        d = parse_date(r.get("date"))
        if d is not None:
            clean_review_dates.append(d)

    stats = {
        "count": len(reviews) if isinstance(reviews, list) else 0,
        "last_review_date": max(clean_review_dates) if clean_review_dates else None,
    }

    nested_scores = doc.get("review_scores", {}) or {}   
    for field in REVIEW_SCORE_FIELDS:
        val = bounded_float(doc.get(field), 0, 100)
        if val is None:
            val = bounded_float(nested_scores.get(field), 0, 100)
        if val is not None:
            stats[field.replace("review_scores_", "")] = val

    return stats   # ← não se pode esquecer isto!


def build_details_doc(doc: dict) -> Optional[dict]:
    payload = {"_id": doc["_id"], "schema_version": 1}
    found = False
    for field in DETAIL_FIELDS:
        value = clean_string(doc.get(field))
        if value is not None:
            payload[field] = value
            found = True
    return payload if found else None

def build_listing_doc(doc: dict) -> dict:
    address = doc.get("address", {}) if isinstance(doc.get("address"), dict) else {}
    availability = doc.get("availability", {}) if isinstance(doc.get("availability"), dict) else {}

    listing = {
        "_id": doc["_id"],
        "listing_url": clean_string(doc.get("listing_url")),
        "name": clean_string(doc.get("name")),
        "property_type": clean_string(doc.get("property_type")),
        "room_type": clean_string(doc.get("room_type")),
        "bed_type": clean_string(doc.get("bed_type")),
        "minimum_nights": parse_int(doc.get("minimum_nights")),
        "maximum_nights": parse_int(doc.get("maximum_nights")),
        "cancellation_policy": clean_string(doc.get("cancellation_policy")),
        "last_scraped": parse_date(doc.get("last_scraped")),
        "calendar_last_scraped": parse_date(doc.get("calendar_last_scraped")),
        "first_review": parse_date(doc.get("first_review")),
        "last_review": parse_date(doc.get("last_review")),
        "accommodates": bounded_int(doc.get("accommodates"), 1, 50),
        "bedrooms": bounded_float(doc.get("bedrooms"), 0, 50),
        "bathrooms": bounded_float(doc.get("bathrooms"), 0, 50),
        "beds": bounded_float(doc.get("beds"), 0, 100),
        "price": parse_decimal128(doc.get("price")),
        "security_deposit": parse_decimal128(doc.get("security_deposit")),
        "cleaning_fee": parse_decimal128(doc.get("cleaning_fee")),
        "extra_people": parse_decimal128(doc.get("extra_people")),
        "guests_included": parse_int(doc.get("guests_included")),
        "number_of_reviews": parse_int(doc.get("number_of_reviews")),
        "amenities": unique_clean_list(doc.get("amenities")),
        "images": {
            "thumbnail_url": clean_string((doc.get("images") or {}).get("thumbnail_url")),
            "medium_url": clean_string((doc.get("images") or {}).get("medium_url")),
            "picture_url": clean_string((doc.get("images") or {}).get("picture_url")),
            "xl_picture_url": clean_string((doc.get("images") or {}).get("xl_picture_url")),
        } if isinstance(doc.get("images"), dict) else {},
        "availability": {
            "availability_30": parse_int(availability.get("availability_30")),
            "availability_60": parse_int(availability.get("availability_60")),
            "availability_90": parse_int(availability.get("availability_90")),
            "availability_365": parse_int(availability.get("availability_365")),
        },
        "address": {
            "street": clean_string(address.get("street")),
            "suburb": clean_string(address.get("suburb")),
            "government_area": clean_string(address.get("government_area")),
            "market": clean_string(address.get("market")),
            "country": clean_string(address.get("country")),
            "country_code": clean_string(address.get("country_code")),
            "location": clean_coords(address.get("location")),
        },
        "host": build_host_snapshot(doc),
        "review_stats": compute_review_stats(doc),
        "details_ref": doc["_id"],
        "schema_version": 1,
    }

    # Consistency fixes
    if listing["first_review"] and listing["last_review"] and listing["first_review"] > listing["last_review"]:
        listing["first_review"], listing["last_review"] = listing["last_review"], listing["first_review"]

    if listing["minimum_nights"] and listing["maximum_nights"]:
        if listing["minimum_nights"] > listing["maximum_nights"]:
            listing["minimum_nights"], listing["maximum_nights"] = listing["maximum_nights"], listing["minimum_nights"]

    return listing

def build_review_docs(doc: dict) -> list[dict]:
    listing_id = doc["_id"]
    out = []
    seen_keys = set()

    for review in doc.get("reviews", []) if isinstance(doc.get("reviews"), list) else []:
        reviewer = review.get("reviewer", {}) if isinstance(review.get("reviewer"), dict) else {}
        comments = clean_string(review.get("comments"))
        review_date = parse_date(review.get("date"))
        reviewer_id = clean_string(review.get("reviewer_id") or reviewer.get("reviewer_id") or reviewer.get("_id"))
        reviewer_name = clean_string(review.get("reviewer_name") or reviewer.get("name"))
        comments_hash = safe_hash_text(comments)

        dedup_key = (
            clean_string(listing_id),
            reviewer_id,
            review_date,
            comments_hash,
        )

        # Drop exact duplicate reviews within the source listing document.
        if dedup_key in seen_keys:
            continue
        seen_keys.add(dedup_key)

        review_id = review.get("_id")
        if is_missing(review_id):
            review_id = safe_hash_text(
                f"{clean_string(listing_id)}|{reviewer_id}|{review_date.isoformat() if review_date else 'nodate'}|{comments_hash or 'notext'}"
            )

        review_doc = {
            "_id": clean_string(review_id),
            "listing_id": listing_id,
            "reviewer_id": reviewer_id,
            "reviewer_name": reviewer_name,
            "date": review_date,
            "comments": comments,
            "comments_hash": comments_hash,
            "schema_version": 1,
        }

        out.append(review_doc)

    return out
```


## 6. Rebuild the target collections

This cell:
- drops the BI-ready target collections
- reads every source listing
- cleans and reshapes data
- writes into `listings`, `listing_details`, `hosts`, and `reviews`

The source collection remains unchanged.


```python
# Drop only target collections, never the source collection.
for collection_name in [
    listings_collection_name,
    listing_details_collection_name,
    hosts_collection_name,
    reviews_collection_name,
]:
    target_database[collection_name].drop()

host_upserts = []
listing_batch = []
details_batch = []
review_batch = []

batch_size = 500

n_source = 0
n_listings = 0
n_details = 0
n_reviews = 0
n_hosts = 0

cursor = source_collection.find({}, no_cursor_timeout=True)

try:
    for doc in cursor:
        n_source += 1

        listing_doc = build_listing_doc(doc)
        details_doc = build_details_doc(doc)
        host_doc = build_host_doc(doc)
        review_docs = build_review_docs(doc)

        listing_batch.append(listing_doc)
        n_listings += 1

        if details_doc is not None:
            details_batch.append(details_doc)
            n_details += 1

        if host_doc is not None:
            host_upserts.append(
                UpdateOne(
                    {"_id": host_doc["_id"]},
                    {"$set": host_doc},
                    upsert=True,
                )
            )
            n_hosts += 1

        if review_docs:
            review_batch.extend(review_docs)
            n_reviews += len(review_docs)

        if len(listing_batch) >= batch_size:
            listings_collection.insert_many(listing_batch, ordered=False)
            listing_batch = []

        if len(details_batch) >= batch_size:
            listing_details_collection.insert_many(details_batch, ordered=False)
            details_batch = []

        if len(review_batch) >= batch_size:
            reviews_collection.insert_many(review_batch, ordered=False)
            review_batch = []

        if len(host_upserts) >= batch_size:
            hosts_collection.bulk_write(host_upserts, ordered=False)
            host_upserts = []

    if listing_batch:
        listings_collection.insert_many(listing_batch, ordered=False)

    if details_batch:
        listing_details_collection.insert_many(details_batch, ordered=False)

    if review_batch:
        reviews_collection.insert_many(review_batch, ordered=False)

    if host_upserts:
        hosts_collection.bulk_write(host_upserts, ordered=False)

finally:
    cursor.close()

{
    "source_documents_read": n_source,
    "listings_written": n_listings,
    "listing_details_written": n_details,
    "reviews_written": n_reviews,
    "host_upserts_applied": n_hosts,
}
```

# C. Confort Analytics Expert

**Modeling / analytical assumptions used**

1. **City field**: we use `address.market` as the city label because this field is the cleanest market-level geographic attribute in the remodeled `listings` collection.

2. **Effective price for two guests staying one week**:

   $$\text{effective\_price\_pppn} = \frac{7 \cdot price + cleaning\_fee + 7 \cdot extra\_people \cdot \max(0,\, 2 - guests\_included)}{14}$$

   where `pppn` means **price per person per night**. Missing monetary fees are treated as zero; missing `guests_included` is treated as 1.

3. **Amenity filtering thresholds**:
   - Amenities appearing in **fewer than 20 properties** in a city are excluded as statistically unreliable (too few observations for a meaningful median comparison).
   - Amenities appearing in **more than 95% of properties** in a city are excluded from premium/value analysis — they offer no meaningful variation between the two groups.

4. **C2**: Properties with at least 30 reviews and a `review_stats.rating` in the top 15% for their city and property type are considered *top-rated*. Amenities appearing in ≥ 85% of top-rated properties are classified as *essential*; the rest are *luxury*.

5. **C3 weighted value metric**: `value_score = 0.6 × normalize(review_scores_value) + 0.4 × normalize(1 − price_per_person_per_night)`.
   - Weight 0.6 on the review value score because it is a direct guest-expressed perception of value and is the most relevant signal.
   - Weight 0.4 on the inverse normalised price because a lower price for the same experience increases perceived value, but we weight it less than the direct review signal.
   - Normalisation is applied per city to make scores comparable within each market.
   - Properties missing either `review_scores_value` or a valid price are excluded.





## C0. Shared assumptions and helper setup

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configure seaborn
sns.set_theme(style="whitegrid", palette="pastel")

# Common projection and price formula (used in C1, C3)
price_calc_pipeline = [
    {"$addFields": {
        "price_num": {"$toDecimal": {"$ifNull": ["$price", 0]}},
        "cleaning_fee_num": {"$toDecimal": {"$ifNull": ["$cleaning_fee", 0]}},
        "extra_people_num": {"$toDecimal": {"$ifNull": ["$extra_people", 0]}},
        "guests_included_num": {"$ifNull": ["$guests_included", 1]}
    }},
    {"$addFields": {
        "extra_guests": {
            "$max": [0, {"$subtract": [2, "$guests_included_num"]}]
        }
    }},
    {"$addFields": {
        "pppn": {
            "$divide": [
                {"$add": [
                    {"$multiply": [7, "$price_num"]},
                    "$cleaning_fee_num",
                    {"$multiply": [7, "$extra_people_num", "$extra_guests"]}
                ]},
                14
            ]
        }
    }}
]

```

## C1. Amenity Premium Analysis

Determine which amenities are associated with higher prices per night in Hong Kong, Montreal, and Barcelona.


```python
# Pipeline for C1
MIN_COUNT    = 20  # exclude amenities with fewer than this many listings in the city
MAX_PREV     = 0.95  # exclude amenities present in >95% of a city's listings (no variation)

docs = list(listings_collection.aggregate([
    {"$match": {"address.market": {"$in": ["Hong Kong", "Montreal", "Barcelona"]}}},
    *price_calc_pipeline,
    {"$project": {
        "_id": 1,
        "market": "$address.market",
        "pppn": {"$toDouble": "$pppn"},
        "amenities": {"$ifNull": ["$amenities", []]}
    }}
]))
df_c1 = pd.DataFrame(docs)
df_c1 = df_c1.dropna(subset=['pppn'])

c1_top_amenities    = {}   # top-3 positive premium per city
c1_bottom_amenities = {}   # top-3 negative premium (price drop) per city
c1_all_stats        = {}   # full stats dataframe per city (used in visualization)

for city in ['Hong Kong', 'Montreal', 'Barcelona']:
    city_df = df_c1[df_c1['market'] == city].copy()
    total   = len(city_df)
    all_amenities = set(am for ams in city_df['amenities'] for am in ams)

    amenity_stats = []
    for am in all_amenities:
        has_am     = city_df['amenities'].apply(lambda x: am in x)
        count      = has_am.sum()
        prevalence = count / total if total > 0 else 0

        # Skip ultra-rare, near-universal amenities, and data errors
        if count < MIN_COUNT or prevalence > MAX_PREV or "translation missing" in am:
            continue


        with_med    = city_df[has_am]['pppn'].median()
        without_med = city_df[~has_am]['pppn'].median()
        amenity_stats.append({
            'amenity':    am,
            'premium':    with_med - without_med,
            'with_med':   with_med,
            'without_med':without_med,
            'count':      count,
            'prevalence': prevalence
        })

    stats_df = pd.DataFrame(amenity_stats).sort_values('premium', ascending=False)
    c1_all_stats[city]        = stats_df
    c1_top_amenities[city]    = stats_df.head(3)['amenity'].tolist()
    c1_bottom_amenities[city] = stats_df.tail(3)['amenity'].tolist()

print("Top-3 POSITIVE premium amenities by city:")
for k, v in c1_top_amenities.items():
    print(f"  {k}: {', '.join(v)}")

print("\nTop-3 NEGATIVE premium amenities (associated with price DROP) by city:")
for k, v in c1_bottom_amenities.items():
    premiums = c1_all_stats[k].set_index('amenity').loc[v, 'premium'].round(2).tolist()
    print(f"  {k}: {', '.join(f'{a} ({p:+.2f})' for a, p in zip(v, premiums))}")
```

```python
# Visualization for C1 — 18 box plots: 3 cities × 3 amenities × 2 groups (with / without)
fig, axes = plt.subplots(3, 3, figsize=(18, 15), sharey=False)
fig.suptitle('Price Premium of Top 3 Amenities per City\n(Price per Person per Night)',
             fontsize=16, y=1.01)

COLORS = {'with': '#4472C4', 'without': '#ED7D31'}

# 1. We create dictionaries to hold custom limits and currencies for each city
Y_LIMITS = {'Hong Kong': 1500, 'Montreal': 350, 'Barcelona': 350}
CURRENCIES = {'Hong Kong': 'HKD', 'Montreal': 'CAD', 'Barcelona': 'EUR'}

for i, market in enumerate(['Hong Kong', 'Montreal', 'Barcelona']):
    city_df = df_c1[df_c1['market'] == market]
    top_3   = c1_top_amenities[market]

    for j, am in enumerate(top_3):
        ax   = axes[i, j]
        mask = city_df['amenities'].apply(lambda x: am in x)

        plot_df = pd.concat([
            city_df[mask].assign(group=f'With\n{am}'),
            city_df[~mask].assign(group='Without')
        ])

        sns.boxplot(
            data=plot_df, x='group', y='pppn',
            palette=[COLORS['with'], COLORS['without']],
            order=[f'With\n{am}', 'Without'],
            showfliers=True,          # show outliers (task says include min/max)
            flierprops=dict(marker='o', markersize=2, alpha=0.3, color='grey'),
            ax=ax
        )

        # Apply the correct currency to titles and labels
        curr = CURRENCIES[market]
        premium = c1_all_stats[market].set_index('amenity').loc[am, 'premium']
        
        ax.set_title(f'{market}\n{am}\n(premium: {premium:+.2f} {curr}/night)', fontsize=9)
        ax.set_xlabel('')
        ax.set_ylabel(f'Price per person per night ({curr})' if j == 0 else '')
        ax.tick_params(axis='x', labelsize=8)
        
        # 2. This sets the perfect zoom level depending on which city is currently plotting
        ax.set_ylim(0, Y_LIMITS[market])

plt.tight_layout()
plt.savefig('c1_amenity_premium.png', dpi=150, bbox_inches='tight')
plt.show()
```

## C2. Luxury Amenity of Top-Rated Properties

Identify the amenities most common among top-rated properties and quantify their prevalence in Sydney and New York.


```python
# Pipeline for C2
prop_types = ['Aparthotel', 'Apartment', 'Condominium', 'Loft', 'Serviced apartment']
markets    = ['Sydney', 'New York']

# Step 1: compute per-city 85th-percentile rating cutoff
cutoffs = {}
for city in markets:
    scores = [
        doc['review_stats']['rating']
        for doc in listings_collection.find(
            {
                "address.market":    city,
                "property_type":     {"$in": prop_types},
                "review_stats.count":{"$gte": 30},
                "review_stats.rating":{"$exists": True, "$ne": None}
            },
            {"review_stats.rating": 1}
        )
        if doc.get('review_stats', {}).get('rating') is not None
    ]
    cutoffs[city] = float(np.percentile(scores, 85)) if scores else 0.0
    print(f"{city}: 85th-percentile cutoff = {cutoffs[city]:.1f}  (n={len(scores)} qualifying listings)")

# Step 2: aggregation pipeline (used for BOTH the results and the explain call)
def build_c2_pipeline(city, cutoff):
    return [
        {"$match": {
            "address.market":     city,
            "property_type":      {"$in": prop_types},
            "review_stats.count": {"$gte": 30},
            "review_stats.rating":{"$gte": cutoff}
        }},
        {"$project": {"amenities": {"$ifNull": ["$amenities", []]}}},
        {"$unwind": "$amenities"},
        {"$group": {"_id": "$amenities", "count": {"$sum": 1}}},
        {"$project": {"amenity": "$_id", "count": 1, "_id": 0}},
        {"$sort": {"count": -1}}
    ]

all_results = []
total_top   = {}

for city in markets:
    cutoff  = cutoffs[city]
    pipeline = build_c2_pipeline(city, cutoff)
    total_top[city] = listings_collection.count_documents({
        "address.market":     city,
        "property_type":      {"$in": prop_types},
        "review_stats.count": {"$gte": 30},
        "review_stats.rating":{"$gte": cutoff}
    })
    print(f"{city}: {total_top[city]} top-rated listings")

    for doc in listings_collection.aggregate(pipeline):
        doc['market']     = city
        doc['prevalence'] = (doc['count'] / total_top[city]) * 100 if total_top[city] > 0 else 0
        all_results.append(doc)

df_c2 = pd.DataFrame(all_results)
df_c2['category'] = np.where(df_c2['prevalence'] >= 85, 'Essential', 'Luxury')

# Top-10 luxury amenities per city
lux_df = df_c2[df_c2['category'] == 'Luxury'].copy()
top_lux_per_city = (
    lux_df.sort_values(['market', 'prevalence'], ascending=[True, False])
          .groupby('market')
          .head(10)
)
c2_top_amenities = top_lux_per_city['amenity'].unique().tolist()
print(f"\nUnique amenities across both cities' top-10 luxury lists: {len(c2_top_amenities)}")
```

```python
# C2 Query Optimization — Before / After comparison table
# --------------------------------------------------------
# Strategy: use collection.find().explain() is not valid for aggregations.
# The correct approach for aggregation explain is target_database.command()
# with explain=True but WITHOUT verbosity (unsupported on community MongoDB).
# For executionStats, we fall back to timing real runs when the explain doc
# does not contain them (which is common on older/community MongoDB versions).

import time
from pymongo import ASCENDING

INDEX_NAME = "c2_market_proptype_count_rating_idx"

def extract_explain_metrics(explain_doc):
    """
    Tries every known layout of the MongoDB aggregate explain output
    and returns a dict with whatever metrics it can find.
    """
    exec_stats = {}
    stage_name = 'N/A'
    index_name_used = 'none (COLLSCAN)'

    # ---- Layout A: top-level executionStats (standalone mongod, newer versions) ----
    if 'executionStats' in explain_doc:
        exec_stats = explain_doc['executionStats']
        qp = explain_doc.get('queryPlanner', {})
        wp = qp.get('winningPlan', {})
        stage_name = wp.get('stage', wp.get('inputStage', {}).get('stage', 'N/A'))
        # Look for index name inside the winning plan tree
        inp = wp.get('inputStage', {})
        index_name_used = inp.get('indexName', 'none (COLLSCAN)')

    # ---- Layout B: stages[0].$cursor (sharded or older) ----
    elif 'stages' in explain_doc:
        stages = explain_doc.get('stages', [])
        if stages:
            cursor_block = stages[0].get('$cursor', {})
            exec_stats   = cursor_block.get('executionStats', {})
            qp  = cursor_block.get('queryPlanner', {})
            wp  = qp.get('winningPlan', {})
            stage_name = wp.get('stage', wp.get('inputStage', {}).get('stage', 'N/A'))
            inp = wp.get('inputStage', {})
            index_name_used = inp.get('indexName', 'none (COLLSCAN)')

    # ---- Layout C: queryPlanner only (no executionStats at all) ----
    # Some community builds return only the query plan without running anything.
    # In this case exec_stats stays empty; we fill it from a real timed run below.
    if not exec_stats:
        stage_name = '(plan only — timed run used)'

    return exec_stats, stage_name, index_name_used


def run_timed_and_explain(pipeline, city):
    """
    1. Gets the explain document (attempts executionStats, falls back to queryPlanner).
    2. Tries to extract executionStats from the explain output.
    3. If executionStats are missing, times a real run as fallback.
    Returns a flat metrics dict.
    """
    # Step 1 — explain with executionStats if supported
    try:
        explain_doc = target_database.command('explain', {'aggregate': listings_collection_name, 'pipeline': pipeline, 'cursor': {}}, verbosity='executionStats')
    except Exception:
        explain_doc = target_database.command('aggregate', listings_collection_name, pipeline=pipeline, explain=True)

    exec_stats, stage_name, index_name_used = extract_explain_metrics(explain_doc)

    # Step 2 — real timed run (always done so we always have a wall-clock time)
    start = time.perf_counter()
    list(listings_collection.aggregate(pipeline))
    elapsed_ms = round((time.perf_counter() - start) * 1000, 1)

    # Prefer MongoDB's own executionTimeMillis if available, else use our timer
    mongo_time = exec_stats.get('executionTimeMillis')
    final_time = mongo_time if mongo_time is not None else elapsed_ms

    return {
        'city':          city,
        'stage':         stage_name,
        'index':         index_name_used,
        'exec_time_ms':  final_time,
        'docs_examined': exec_stats.get('totalDocsExamined',  'N/A'),
        'keys_examined': exec_stats.get('totalKeysExamined',  'N/A'),
        'docs_returned': exec_stats.get('nReturned',          'N/A'),
    }


# ══════════════════════════════════════════════════════════
# BEFORE — drop the index so we get a clean baseline scan
# ══════════════════════════════════════════════════════════
try:
    listings_collection.drop_index(INDEX_NAME)
    print(f"Dropped existing '{INDEX_NAME}' for clean baseline.")
except Exception:
    print("Index did not exist yet — baseline is already a clean scan.")

before_rows = []
for city in markets:
    row = run_timed_and_explain(build_c2_pipeline(city, cutoffs[city]), city)
    row['phase'] = 'BEFORE'
    before_rows.append(row)
    print(f"[BEFORE] {city}: {row['exec_time_ms']} ms | stage={row['stage']}")

# ══════════════════════════════════════════════════════════
# CREATE the compound index
# ══════════════════════════════════════════════════════════
listings_collection.create_index(
    [
        ("address.market",       ASCENDING),
        ("property_type",        ASCENDING),
        ("review_stats.count",   ASCENDING),
        ("review_stats.rating",  ASCENDING),
    ],
    name=INDEX_NAME
)
print(f"\nIndex '{INDEX_NAME}' created.")

# ══════════════════════════════════════════════════════════
# AFTER — re-run with the index in place
# ══════════════════════════════════════════════════════════
after_rows = []
for city in markets:
    row = run_timed_and_explain(build_c2_pipeline(city, cutoffs[city]), city)
    row['phase'] = 'AFTER'
    after_rows.append(row)
    print(f"[AFTER]  {city}: {row['exec_time_ms']} ms | stage={row['stage']}")

# ══════════════════════════════════════════════════════════
# Render comparison table
# ══════════════════════════════════════════════════════════
df_compare = (
    pd.DataFrame(before_rows + after_rows)
      .set_index(['phase', 'city'])
      [['index', 'stage', 'exec_time_ms', 'docs_examined', 'keys_examined', 'docs_returned']]
)
df_compare.columns = ['Index Used', 'Stage', 'Exec Time (ms)', 
                       'Docs Examined', 'Keys Examined', 'Docs Returned']

print("\n=== C2 Query Optimization: Before vs After ===")
display(df_compare)
```

## C3. Amenity Value Analysis and Suggestion

Analyze which amenities are associated with high guest-perceived value in Hong Kong and Barcelona.


```python
# Pipeline for C3
MIN_COUNT_C3 = 20    # minimum listings that must have an amenity for it to be considered
MAX_PREV_C3  = 0.85  # exclude near-universal amenities (>85% prevalence)

c3_pipeline = [
    {"$match": {"address.market": {"$in": ["Hong Kong", "Barcelona"]}}},
    *price_calc_pipeline,
    {"$project": {
        "_id": 1,
        "market":       "$address.market",
        "pppn":         {"$toDouble": "$pppn"},
        "review_value": {"$toDouble": "$review_stats.value"},
        "amenities":    {"$ifNull": ["$amenities", []]}
    }}
]

df_c3 = pd.DataFrame(list(listings_collection.aggregate(c3_pipeline)))
df_c3 = df_c3.dropna(subset=['pppn', 'review_value'])
df_c3 = df_c3[df_c3['pppn'] > 0].copy()   # drop zero-price entries

# Per-city min-max normalisation and value score
for city in ["Hong Kong", "Barcelona"]:
    mask = df_c3['market'] == city
    if df_c3[mask].empty:
        continue
    p_min, p_max = df_c3[mask]['pppn'].min(),          df_c3[mask]['pppn'].max()
    v_min, v_max = df_c3[mask]['review_value'].min(),  df_c3[mask]['review_value'].max()
    p_denom = p_max - p_min if p_max > p_min else 1
    v_denom = v_max - v_min if v_max > v_min else 1

    df_c3.loc[mask, 'norm_val']       = (df_c3.loc[mask, 'review_value'] - v_min) / v_denom
    df_c3.loc[mask, 'inv_norm_price'] = 1 - ((df_c3.loc[mask, 'pppn'] - p_min) / p_denom)
    df_c3.loc[mask, 'value_score']    = (
        0.6 * df_c3.loc[mask, 'norm_val'] +
        0.4 * df_c3.loc[mask, 'inv_norm_price']
    )

# Amenity association analysis with MIN_COUNT + MAX_PREV filters
c3_top_amenities = {}
c3_am_stats_all  = {}

for city in ["Hong Kong", "Barcelona"]:
    city_df = df_c3[df_c3['market'] == city].copy()
    if city_df.empty:
        continue
    total = len(city_df)
    all_amenities = set(am for ams in city_df['amenities'] for am in ams)

    am_stats = []
    for am in all_amenities:
        if "translation missing" in am:
            continue
        has_am     = city_df['amenities'].apply(lambda x: am in x)
        count      = has_am.sum()
        prevalence = count / total

        # Apply both filters
        if count < MIN_COUNT_C3:
            continue
        if prevalence > MAX_PREV_C3:
            continue

        with_med    = city_df[has_am]['value_score'].median()
        without_med = city_df[~has_am]['value_score'].median()
        am_stats.append({
            'amenity':     am,
            'val_premium': with_med - without_med,
            'with_med':    with_med,
            'without_med': without_med,
            'count':       count,
            'prevalence':  round(prevalence * 100, 1)
        })

    stats_df = pd.DataFrame(am_stats).sort_values('val_premium', ascending=False)
    c3_am_stats_all[city]  = stats_df
    c3_top_amenities[city] = stats_df.head(3)['amenity'].tolist()

print("Top 3 value-driving amenities by city:")
for city, amenities in c3_top_amenities.items():
    rows = c3_am_stats_all[city].set_index('amenity').loc[amenities]
    for am, row in rows.iterrows():
        print(f"  [{city}] {am}  |  val_premium={row['val_premium']:+.4f}  "
              f"|  prevalence={row['prevalence']}%  |  n={int(row['count'])}")
```

```python
# C3 — Suggestion logic: identify below-median properties and recommend missing top-3 amenities

df_c3['suggested_amenities'] = None   # initialise column

for city in ["Hong Kong", "Barcelona"]:
    mask    = df_c3['market'] == city
    city_df = df_c3[mask].copy()
    if city_df.empty:
        continue

    median_score = city_df['value_score'].median()
    top_3_set    = set(c3_top_amenities[city])

    low_mask = mask & (df_c3['value_score'] < median_score)

    def suggest_for_city(amenities):
        missing = top_3_set - set(amenities)
        return sorted(missing) if missing else []

    df_c3.loc[low_mask, 'suggested_amenities'] = (
        df_c3.loc[low_mask, 'amenities'].apply(suggest_for_city)
    )

# Display summary
sample = df_c3[df_c3['suggested_amenities'].apply(
    lambda x: isinstance(x, list) and len(x) > 0
)][['market', 'value_score', 'amenities', 'suggested_amenities']]

print(f"Properties below median value score with actionable suggestions: {len(sample)}")
print("\nSample output (first 10 rows):")
display(sample[['market', 'value_score', 'suggested_amenities']].head(10).reset_index(drop=True))

# Count suggestions per city and per amenity
print("\nSuggestion frequency (how often each amenity is recommended per city):")
for city in ["Hong Kong", "Barcelona"]:
    city_sample = sample[sample['market'] == city]
    from collections import Counter
    counts = Counter(am for lst in city_sample['suggested_amenities'] for am in lst)
    print(f"\n  {city}:")
    for am, cnt in counts.most_common():
        print(f"    {am}: recommended to {cnt} properties")
```

The weighted value metric (`0.6 × normalized review_value + 0.4 × normalized inverse price`) captures both guest-expressed satisfaction and price competitiveness. In terms of value-driving amenities (evaluating 455 properties total), our weighted scoring model reveals distinct priorities per market:

## D0. Shared assumptions and helper setup

These cells implement **D1**, **D2**, and **D3** on top of the BI-ready collections created above.

**Modeling / analytical assumptions used**

1. **City field**: we use `address.market` as the city label because this field is the cleanest market-level geographic attribute in the remodeled `listings` collection.
2. **Professional host**: a host is classified as professional if the host manages **more than one listing in the full remodeled `listings` collection**, not only inside a single city.
3. **Effective price for two guests staying one week**:

   \[
   effective_price_pppn = (7.price + cleaning_fee + 7. extra_people.max(0,2-guests_included))/14
   \]

   where `pppn` means **price per person per night**.
4. Missing monetary fees are treated as zero; missing `guests_included` is treated as 1.
5. In D2, host satisfaction is measured at **host level** by a **weighted average review score**, where the weight is the number of reviews attached to each listing. This reduces the influence of hosts whose average is driven by only a few reviews.
6. Binary feature definitions used in D2:
   - `is_superhost`: `host_is_superhost == True`
   - `is_fast_responder`: `host_response_time` in `{"within an hour", "within a few hours"}`
   - `is_verified`: `host_identity_verified == True`
   - `high_response_rate`: response rate (90%)
   - `relaxed_cancellation`: at least 50% of the host's listings use `flexible` or `moderate`
7. In D3, popularity is proxied by **future unavailability**, with more weight on nearer horizons:

   popularity_score = 100[
   0.40(1-a_30/30) +
   0.30(1-a_60/60 +
   0.20(1-a_90/90) +
   0.10(1-a_365/365)
   ]
   

   This yields a score on a roughly 0–100 scale, where higher means more booked/unavailable.
8. The popularity analysis in D3 is restricted to **professional hosts**, because the task states that for professionally rented properties unavailability can be interpreted as demand/bookings.
9. Satisfaction categories in D3 are split using **tertiles** of `review_scores_rating` among professional-host listings with valid popularity and valid rating. This gives balanced low / medium / high groups for comparison.


```python
from IPython.display import Markdown, display

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

TARGET_CITIES = ["Hong Kong", "Montreal", "Barcelona"]
HOST_TYPE_ORDER = ["Professional", "Amateur"]

def d128_to_float(x):
    if x is None:
        return np.nan
    if isinstance(x, (int, float)):
        return float(x)
    if hasattr(x, "to_decimal"):
        return float(x.to_decimal())
    try:
        return float(x)
    except Exception:
        return np.nan

def safe_numeric(x, default=np.nan):
    try:
        if x is None:
            return default
        return float(x)
    except Exception:
        return default

def pct_string_to_float(x):
    if x is None:
        return np.nan
    if isinstance(x, (int, float)):
        return float(x)
    s = str(x).strip().replace("%", "")
    if s == "":
        return np.nan
    try:
        return float(s)
    except Exception:
        return np.nan

def summarize_distribution(df, value_col, group_cols):
    out = (
        df.groupby(group_cols)[value_col]
          .agg(
              n="count",
              min_value="min",
              q1=lambda s: s.quantile(0.25),
              median="median",
              q3=lambda s: s.quantile(0.75),
              max_value="max",
              mean="mean",
          )
          .reset_index()
    )
    out["iqr"] = out["q3"] - out["q1"]
    return out

def recursive_find_values(obj, key):
    hits = []
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == key:
                hits.append(v)
            hits.extend(recursive_find_values(v, key))
    elif isinstance(obj, list):
        for item in obj:
            hits.extend(recursive_find_values(item, key))
    return hits

def flatten_stage_names(obj):
    stages = []
    if isinstance(obj, dict):
        stage_name = obj.get("stage")
        if stage_name is not None:
            stages.append(stage_name)
        for v in obj.values():
            stages.extend(flatten_stage_names(v))
    elif isinstance(obj, list):
        for item in obj:
            stages.extend(flatten_stage_names(item))
    return list(dict.fromkeys(stages))

def extract_explain_metrics(explain_doc, label):
    exec_stats = explain_doc.get("executionStats", {}) if isinstance(explain_doc, dict) else {}
    docs_examined = recursive_find_values(explain_doc, "totalDocsExamined")
    keys_examined = recursive_find_values(explain_doc, "totalKeysExamined")
    mem_usage = recursive_find_values(explain_doc, "maxAccumulatorMemoryUsageBytes")
    used_disk = recursive_find_values(explain_doc, "usedDisk")

    return {
        "version": label,
        "execution_time_ms": exec_stats.get("executionTimeMillis"),
        "n_returned": exec_stats.get("nReturned"),
        "total_docs_examined": max(docs_examined) if docs_examined else None,
        "total_keys_examined": max(keys_examined) if keys_examined else None,
        "stages": ", ".join(flatten_stage_names(explain_doc)),
        "used_disk": True if any(used_disk) else False if used_disk else None,
        "max_accumulator_memory_bytes": max(mem_usage) if mem_usage else None,
    }

def aggregate_explain(collection, pipeline):
    try:
        return target_database.command(
            "aggregate",
            collection.name,
            pipeline=pipeline,
            explain=True,
            cursor={},
            allowDiskUse=True,
        )
    except Exception as e:
        return {"error": str(e)}
```

```python

# D1A. Build the analysis dataset for the three target cities

host_listing_count_map = {
    row["_id"]: row["n_listings"]
    for row in listings_collection.aggregate([
        {"$match": {"host.host_id": {"$ne": None}}},
        {"$group": {"_id": "$host.host_id", "n_listings": {"$sum": 1}}},
    ])
}

d1_pipeline = [
    {
        "$match": {
            "address.market": {"$in": TARGET_CITIES},
            "host.host_id": {"$ne": None},
            "price": {"$ne": None},
        }
    },
    {
        "$project": {
            "_id": 1,
            "city": "$address.market",
            "host_id": "$host.host_id",
            "price": 1,
            "cleaning_fee": 1,
            "extra_people": 1,
            "guests_included": 1,
            "property_type": 1,
            "room_type": 1,
        }
    }
]

d1_df = pd.DataFrame(list(listings_collection.aggregate(d1_pipeline)))
d1_df["price"] = d1_df["price"].apply(d128_to_float)
d1_df["cleaning_fee"] = d1_df["cleaning_fee"].apply(lambda x: d128_to_float(x) if x is not None else 0.0).fillna(0.0)
d1_df["extra_people"] = d1_df["extra_people"].apply(lambda x: d128_to_float(x) if x is not None else 0.0).fillna(0.0)
d1_df["guests_included"] = pd.to_numeric(d1_df["guests_included"], errors="coerce").fillna(1).clip(lower=0)

extra_guest_count = np.maximum(0, 2 - d1_df["guests_included"])
d1_df["effective_price_pppn"] = (
    7 * d1_df["price"].fillna(0.0)
    + d1_df["cleaning_fee"]
    + 7 * d1_df["extra_people"] * extra_guest_count
) / 14.0

d1_df["global_host_listing_count"] = d1_df["host_id"].map(host_listing_count_map)
d1_df["host_type"] = np.where(d1_df["global_host_listing_count"] > 1, "Professional", "Amateur")

d1_df = d1_df[
    d1_df["effective_price_pppn"].notna()
    & np.isfinite(d1_df["effective_price_pppn"])
    & (d1_df["effective_price_pppn"] > 0)
].copy()

d1_df.head()
```

## D2. Host Features Impact on Satisfaction

```python
# D2A. Baseline raw pipeline: compute host-level satisfaction directly from listings + hosts
# More robust version with diagnostics and type-safe join

d2_diagnostic = {
    "listings_total": listings_collection.count_documents({}),
    "listings_with_host_id": listings_collection.count_documents({"host.host_id": {"$ne": None}}),
    "listings_with_review_rating": listings_collection.count_documents({"review_stats.rating": {"$ne": None}}),
    "hosts_total": hosts_collection.count_documents({}),
}

display(pd.DataFrame([d2_diagnostic]))

host_feature_raw_pipeline = [
    {
        "$match": {
            "host.host_id": {"$ne": None},
            "review_stats.rating": {"$ne": None},
        }
    },
    {
        "$addFields": {
            "host_id_str": {"$toString": "$host.host_id"},
            "rating_num": {"$toDouble": "$review_stats.rating"},
            "rating_weight": {
                "$cond": [
                    {"$gt": [{"$ifNull": ["$review_stats.count", 0]}, 0]},
                    {"$toDouble": {"$ifNull": ["$review_stats.count", 1]}},
                    1.0,
                ]
            },
        }
    },
    {
        "$lookup": {
            "from": hosts_collection_name,
            "let": {"hid": "$host_id_str"},
            "pipeline": [
                {
                    "$match": {
                        "$expr": {
                            "$eq": ["$_id", "$$hid"]
                        }
                    }
                }
            ],
            "as": "host_meta",
        }
    },
    {
        "$match": {
            "host_meta.0": {"$exists": True}
        }
    },
    {
        "$unwind": "$host_meta"
    },
    {
        "$project": {
            "_id": 0,
            "host_id": "$host_id_str",
            "rating": "$rating_num",
            "rating_weight": 1,
            "is_superhost": {
                "$cond": [
                    {"$eq": ["$host_meta.host_is_superhost", True]},
                    1,
                    0,
                ]
            },
            "is_verified": {
                "$cond": [
                    {"$eq": ["$host_meta.host_identity_verified", True]},
                    1,
                    0,
                ]
            },
            "is_fast_responder": {
                "$cond": [
                    {
                        "$in": [
                            {"$toLower": {"$ifNull": ["$host_meta.host_response_time", ""]}},
                            ["within an hour", "within a few hours"],
                        ]
                    },
                    1,
                    0,
                ]
            },
            "high_response_rate": {
                "$cond": [
                    {
                        "$gte": [
                            {
                                "$convert": {
                                    "input": {
                                        "$replaceAll": {
                                            "input": {"$ifNull": ["$host_meta.host_response_rate", "0%"]},
                                            "find": "%",
                                            "replacement": "",
                                        }
                                    },
                                    "to": "int",
                                    "onError": 0,
                                    "onNull": 0,
                                }
                            },
                            90,
                        ]
                    },
                    1,
                    0,
                ]
            },
            "relaxed_cancel": {
                "$cond": [
                    {
                        "$in": [
                            {"$toLower": {"$ifNull": ["$cancellation_policy", ""]}},
                            ["flexible", "moderate"],
                        ]
                    },
                    1,
                    0,
                ]
            },
        }
    },
    {
        "$group": {
            "_id": "$host_id",
            "weighted_sum_rating": {"$sum": {"$multiply": ["$rating", "$rating_weight"]}},
            "total_weight": {"$sum": "$rating_weight"},
            "n_listings": {"$sum": 1},
            "is_superhost": {"$max": "$is_superhost"},
            "is_verified": {"$max": "$is_verified"},
            "is_fast_responder": {"$max": "$is_fast_responder"},
            "high_response_rate": {"$max": "$high_response_rate"},
            "relaxed_cancel_share": {"$avg": "$relaxed_cancel"},
        }
    },
    {
        "$project": {
            "_id": 1,
            "host_avg_rating": {
                "$cond": [
                    {"$gt": ["$total_weight", 0]},
                    {"$divide": ["$weighted_sum_rating", "$total_weight"]},
                    None,
                ]
            },
            "rating_weight": "$total_weight",
            "n_listings": 1,
            "is_superhost": 1,
            "is_verified": 1,
            "is_fast_responder": 1,
            "high_response_rate": 1,
            "relaxed_cancellation": {
                "$cond": [
                    {"$gte": ["$relaxed_cancel_share", 0.5]},
                    1,
                    0,
                ]
            },
        }
    },
    {
        "$match": {
            "host_avg_rating": {"$ne": None},
            "rating_weight": {"$gt": 0},
        }
    },
]

host_satisfaction_summary_docs = list(
    listings_collection.aggregate(host_feature_raw_pipeline, allowDiskUse=True)
)

print("Host-level rows created:", len(host_satisfaction_summary_docs))

host_satisfaction_summary = pd.DataFrame(host_satisfaction_summary_docs)
display(host_satisfaction_summary.head())

if host_satisfaction_summary.empty:
    raise ValueError(
        "D2A returned zero host-level rows. Check the diagnostic table above and confirm "
        "that listings have host.host_id and review_stats.rating, and that hosts_collection "
        "contains matching _id values."
    )
```

```python
# D2B. Materialize an optimized summary collection (Computed Pattern for repeated analytical reads)
# Use $merge so the collection can be incrementally refreshed instead of dropped and fully rebuilt.

host_satisfaction_summary_collection_name = "host_satisfaction_summary"
host_satisfaction_summary_collection = target_database[host_satisfaction_summary_collection_name]

merge_pipeline = host_feature_raw_pipeline + [
    {
        "$set": {
            "schema_version": 1,
            "last_refreshed_at": "$$NOW",
        }
    },
    {
        "$merge": {
            "into": host_satisfaction_summary_collection_name,
            "on": "_id",
            "whenMatched": "replace",
            "whenNotMatched": "insert",
        }
    },
]

list(listings_collection.aggregate(merge_pipeline, allowDiskUse=True))

# Supportive indexes for host-level analytical slicing
host_satisfaction_summary_collection.create_index([("is_superhost", ASCENDING)], name="is_superhost_idx")
host_satisfaction_summary_collection.create_index([("is_verified", ASCENDING)], name="is_verified_idx")
host_satisfaction_summary_collection.create_index([("is_fast_responder", ASCENDING)], name="is_fast_responder_idx")
host_satisfaction_summary_collection.create_index([("high_response_rate", ASCENDING)], name="high_response_rate_idx")
host_satisfaction_summary_collection.create_index([("relaxed_cancellation", ASCENDING)], name="relaxed_cancellation_idx")
host_satisfaction_summary_collection.create_index([("host_avg_rating", DESCENDING)], name="host_avg_rating_idx")
host_satisfaction_summary_collection.create_index([("n_listings", DESCENDING)], name="n_listings_idx")

print("Rows in host_satisfaction_summary_collection:", host_satisfaction_summary_collection.count_documents({}))
display(pd.DataFrame(list(host_satisfaction_summary_collection.find({}, {"_id": 1, "host_avg_rating": 1, "is_superhost": 1, "last_refreshed_at": 1}).limit(5))))
```

```python
# D2C. Weighted satisfaction scores for each binary attribute + double lollipop chart

d2_feature_map = {
    "is_superhost": "Superhost",
    "is_fast_responder": "Fast responder",
    "is_verified": "Verified host",
    "relaxed_cancellation": "Relaxed cancellation",
    "high_response_rate": "High response rate",
}

summary_count = host_satisfaction_summary_collection.count_documents({})
print("Rows available in host_satisfaction_summary_collection:", summary_count)

if summary_count == 0:
    raise ValueError(
        "host_satisfaction_summary_collection is empty, so D2 scores cannot be computed. "
        "Run D2A and D2B again and inspect the diagnostic output."
    )

d2_rows = []
for field, label in d2_feature_map.items():
    pipeline = [
        {
            "$match": {
                field: {"$in": [0, 1]},
                "host_avg_rating": {"$ne": None},
                "rating_weight": {"$gt": 0},
            }
        },
        {
            "$group": {
                "_id": f"${field}",
                "weighted_sum": {"$sum": {"$multiply": ["$host_avg_rating", "$rating_weight"]}},
                "weight": {"$sum": "$rating_weight"},
                "n_hosts": {"$sum": 1},
            }
        },
        {
            "$project": {
                "_id": 0,
                "feature_field": field,
                "feature_label": label,
                "feature_value": "$_id",
                "score": {
                    "$cond": [
                        {"$gt": ["$weight", 0]},
                        {"$divide": ["$weighted_sum", "$weight"]},
                        None,
                    ]
                },
                "n_hosts": 1,
                "weight": 1,
            }
        },
        {"$sort": {"feature_value": 1}},
    ]
    rows = list(host_satisfaction_summary_collection.aggregate(pipeline))
    d2_rows.extend(rows)

d2_scores = pd.DataFrame(d2_rows)
display(d2_scores)

if d2_scores.empty or "feature_value" not in d2_scores.columns:
    raise ValueError(
        "D2 score aggregation returned no rows. Inspect host_satisfaction_summary_collection "
        "and verify that the binary feature columns exist and contain 0/1 values."
    )

d2_scores["feature_value_label"] = d2_scores["feature_value"].map({0: "No", 1: "Yes"})
display(d2_scores.sort_values(["feature_label", "feature_value"]))

plot_order = list(d2_feature_map.values())
plot_df = d2_scores.copy()
plot_df["feature_label"] = pd.Categorical(plot_df["feature_label"], categories=plot_order, ordered=True)
plot_df = plot_df.sort_values(["feature_label", "feature_value"])

# Keep only features where both No and Yes exist
valid_features = (
    plot_df.groupby("feature_label")["feature_value"]
           .nunique()
           .reset_index(name="n_values")
)
valid_features = valid_features.loc[valid_features["n_values"] == 2, "feature_label"].tolist()

plot_df = plot_df[plot_df["feature_label"].isin(valid_features)].copy()

if plot_df.empty:
    raise ValueError(
        "None of the host attributes has both binary groups present. "
        "You cannot build a double lollipop chart from one-sided categories."
    )

plot_order = [x for x in plot_order if x in valid_features]

x = np.arange(len(plot_order))
y_no = plot_df[plot_df["feature_value"] == 0].sort_values("feature_label")["score"].values
y_yes = plot_df[plot_df["feature_value"] == 1].sort_values("feature_label")["score"].values

fig, ax = plt.subplots(figsize=(12, 6))

for i in range(len(plot_order)):
    ax.plot([x[i], x[i]], [y_no[i], y_yes[i]], color="lightgray", linewidth=2, zorder=1)

ax.scatter(x - 0.05, y_no, s=110, label="No", zorder=2)
ax.scatter(x + 0.05, y_yes, s=110, label="Yes", zorder=3)

for i in range(len(plot_order)):
    ax.text(x[i] - 0.08, y_no[i], f"{y_no[i]:.2f}", va="bottom", ha="right", fontsize=9)
    ax.text(x[i] + 0.08, y_yes[i], f"{y_yes[i]:.2f}", va="bottom", ha="left", fontsize=9)

ax.set_xticks(x)
ax.set_xticklabels(plot_order, rotation=15, ha="right")
ax.set_ylabel("Weighted average satisfaction score")
ax.set_xlabel("Host attribute")
ax.set_title("D2. Weighted satisfaction by host attribute")
ax.grid(axis="y", linestyle="--", alpha=0.35)
ax.legend()
plt.tight_layout()
plt.show()

d2_score_diff = (
    d2_scores.pivot(index="feature_label", columns="feature_value_label", values="score")
             .reset_index()
)

if {"Yes", "No"}.issubset(d2_score_diff.columns):
    d2_score_diff["difference_yes_minus_no"] = d2_score_diff["Yes"] - d2_score_diff["No"]

display(d2_score_diff.sort_values("difference_yes_minus_no", ascending=False))
```

```python
PIPELINE_STAGE_KEYS = {
    "$match", "$group", "$project", "$facet", "$lookup", "$unwind",
    "$sort", "$limit", "$skip", "$set", "$addFields", "$merge",
    "$count", "$replaceRoot", "$replaceWith"
}

def feature_score_facet_pipeline(prefix_pipeline, feature_map=None):
    """
    Build one comparable aggregation pipeline for performance benchmarking.

    The input pipeline must output host-level documents with:
      - host_avg_rating
      - rating_weight
      - binary feature fields in d2_feature_map

    We then compute the weighted satisfaction score for each feature inside
    a $facet so the raw and computed-summary versions do equivalent work.
    """
    if feature_map is None:
        feature_map = d2_feature_map

    facet = {}
    for field, label in feature_map.items():
        facet[field] = [
            {
                "$match": {
                    field: {"$in": [0, 1]},
                    "host_avg_rating": {"$ne": None},
                    "rating_weight": {"$gt": 0},
                }
            },
            {
                "$group": {
                    "_id": f"${field}",
                    "weighted_sum": {
                        "$sum": {"$multiply": ["$host_avg_rating", "$rating_weight"]}
                    },
                    "weight": {"$sum": "$rating_weight"},
                    "n_hosts": {"$sum": 1},
                }
            },
            {
                "$project": {
                    "_id": 0,
                    "feature_field": field,
                    "feature_label": label,
                    "feature_value": "$_id",
                    "score": {
                        "$cond": [
                            {"$gt": ["$weight", 0]},
                            {"$divide": ["$weighted_sum", "$weight"]},
                            None,
                        ]
                    },
                    "n_hosts": 1,
                    "weight": 1,
                }
            },
            {"$sort": {"feature_value": 1}},
        ]

    return list(prefix_pipeline) + [{"$facet": facet}]

def aggregate_explain(collection, pipeline):
    return collection.database.command(
        "explain",
        {
            "aggregate": collection.name,
            "pipeline": pipeline,
            "cursor": {},
            "allowDiskUse": True
        },
        verbosity="executionStats"
    )

def walk_tree(obj):
    if isinstance(obj, dict):
        yield obj
        for v in obj.values():
            yield from walk_tree(v)
    elif isinstance(obj, list):
        for item in obj:
            yield from walk_tree(item)

def collect_stage_names(explain_doc):
    names = []
    for node in walk_tree(explain_doc):
        if not isinstance(node, dict):
            continue

        if "stage" in node and isinstance(node["stage"], str):
            names.append(node["stage"])

        for k in node.keys():
            if k in PIPELINE_STAGE_KEYS:
                names.append(k)

    seen = set()
    ordered = []
    for x in names:
        if x not in seen:
            seen.add(x)
            ordered.append(x)
    return ordered

def collect_numeric_metric(explain_doc, key):
    values = []
    for node in walk_tree(explain_doc):
        if key in node:
            v = node[key]
            if isinstance(v, Number):
                values.append(float(v))
    return max(values) if values else None

def collect_memory_metric(explain_doc):
    values = []
    for node in walk_tree(explain_doc):
        if "maxAccumulatorMemoryUsageBytes" in node:
            v = node["maxAccumulatorMemoryUsageBytes"]
            if isinstance(v, Number):
                values.append(float(v))
            elif isinstance(v, dict):
                values.extend(
                    float(x) for x in v.values() if isinstance(x, Number)
                )
    return max(values) if values else None

def collect_used_disk(explain_doc):
    flags = []
    for node in walk_tree(explain_doc):
        if "usedDisk" in node and isinstance(node["usedDisk"], bool):
            flags.append(node["usedDisk"])
    return any(flags) if flags else None

def benchmark_runtime(collection, pipeline, runs=5):
    list(collection.aggregate(pipeline, allowDiskUse=True))  # warm-up

    times = []
    output_docs = None

    for _ in range(runs):
        t0 = perf_counter()
        result = list(collection.aggregate(pipeline, allowDiskUse=True))
        elapsed_ms = (perf_counter() - t0) * 1000
        times.append(elapsed_ms)

        if output_docs is None:
            output_docs = len(result)

    return {
        "execution_time_ms_avg": round(float(np.mean(times)), 2),
        "execution_time_ms_std": round(float(np.std(times)), 2),
        "execution_time_ms_min": round(float(np.min(times)), 2),
        "execution_time_ms_max": round(float(np.max(times)), 2),
        "execution_time_ms_runs": [round(float(x), 2) for x in times],
        "output_docs": output_docs,
    }

def benchmark_pipeline(collection, pipeline, label, runs=5):
    runtime_stats = benchmark_runtime(collection, pipeline, runs=runs)
    explain_doc = aggregate_explain(collection, pipeline)

    return {
        "version": label,
        **runtime_stats,
        "total_docs_examined": collect_numeric_metric(explain_doc, "totalDocsExamined"),
        "total_keys_examined": collect_numeric_metric(explain_doc, "totalKeysExamined"),
        "used_disk": collect_used_disk(explain_doc),
        "max_accumulator_memory_bytes": collect_memory_metric(explain_doc),
        "stages": " -> ".join(collect_stage_names(explain_doc)),
    }

# Baseline explain on raw listing-level pipeline
d2_raw_score_pipeline = feature_score_facet_pipeline(host_feature_raw_pipeline)
d2_perf_raw_before = benchmark_pipeline(
    listings_collection,
    d2_raw_score_pipeline,
    "Raw pipeline (before extra index)",
    runs=5
)

try:
    listings_collection.create_index(
        [("host.host_id", ASCENDING), ("address.market", ASCENDING), ("review_stats.rating", DESCENDING)],
        name="host_city_rating_idx"
    )
except Exception:
    pass

d2_perf_raw_after = benchmark_pipeline(
    listings_collection,
    d2_raw_score_pipeline,
    "Raw pipeline (after supporting index)",
    runs=5
)

d2_summary_prefix_pipeline = [
    {
        "$match": {
            "host_avg_rating": {"$ne": None},
            "rating_weight": {"$gt": 0},
        }
    }
]
d2_summary_score_pipeline = feature_score_facet_pipeline(d2_summary_prefix_pipeline)

d2_perf_summary = benchmark_pipeline(
    host_satisfaction_summary_collection,
    d2_summary_score_pipeline,
    "Computed host summary pipeline",
    runs=5
)

d2_performance_table = pd.DataFrame([
    d2_perf_raw_before,
    d2_perf_raw_after,
    d2_perf_summary,
])[[
    "version",
    "execution_time_ms_avg",
    "execution_time_ms_std",
    "execution_time_ms_min",
    "execution_time_ms_max",
    "total_docs_examined",
    "total_keys_examined",
    "used_disk",
    "max_accumulator_memory_bytes",
    "output_docs",
    "stages",
    "execution_time_ms_runs",
]]

for col in [
    "total_docs_examined",
    "total_keys_examined",
    "used_disk",
    "max_accumulator_memory_bytes",
]:
    d2_performance_table[col] = d2_performance_table[col].apply(
        lambda x: "N/A" if pd.isna(x) else x
    )

display(d2_performance_table)
```

## D3. Property Popularity and Satisfaction

```python
# D3A. Compute and persist popularity_score in the database for professional-host listings
# Identify professional hosts server-side via the materialized host summary collection
# rather than passing a Python list into $in.

d3_pipeline = [
    {
        "$match": {
            "host.host_id": {"$ne": None},
            "review_stats.rating": {"$ne": None},
            "property_type": {"$ne": None},
        }
    },
    {
        "$addFields": {
            "host_id_str": {"$toString": "$host.host_id"},
        }
    },
    {
        "$lookup": {
            "from": host_satisfaction_summary_collection_name,
            "localField": "host_id_str",
            "foreignField": "_id",
            "as": "host_perf",
        }
    },
    {
        "$unwind": "$host_perf"
    },
    {
        "$match": {
            "host_perf.n_listings": {"$gt": 1}
        }
    },
    {
        "$project": {
            "_id": 1,
            "host_id": "$host_id_str",
            "property_type": 1,
            "review_score": {"$toDouble": "$review_stats.rating"},
            "availability_30": {"$ifNull": ["$availability.availability_30", 30]},
            "availability_60": {"$ifNull": ["$availability.availability_60", 60]},
            "availability_90": {"$ifNull": ["$availability.availability_90", 90]},
            "availability_365": {"$ifNull": ["$availability.availability_365", 365]},
            "price": 1,
            "cleaning_fee": 1,
            "extra_people": 1,
            "guests_included": 1,
        }
    }
]

d3_df = pd.DataFrame(list(listings_collection.aggregate(d3_pipeline, allowDiskUse=True)))
d3_df["price"] = d3_df["price"].apply(d128_to_float)
d3_df["cleaning_fee"] = d3_df["cleaning_fee"].apply(lambda x: d128_to_float(x) if x is not None else 0.0).fillna(0.0)
d3_df["extra_people"] = d3_df["extra_people"].apply(lambda x: d128_to_float(x) if x is not None else 0.0).fillna(0.0)
d3_df["guests_included"] = pd.to_numeric(d3_df["guests_included"], errors="coerce").fillna(1).clip(lower=0)

for col, horizon in [("availability_30", 30), ("availability_60", 60), ("availability_90", 90), ("availability_365", 365)]:
    d3_df[col] = pd.to_numeric(d3_df[col], errors="coerce").fillna(horizon).clip(lower=0, upper=horizon)

d3_df["effective_price_pppn"] = (
    7 * d3_df["price"].fillna(0.0)
    + d3_df["cleaning_fee"]
    + 7 * d3_df["extra_people"] * np.maximum(0, 2 - d3_df["guests_included"])
) / 14.0

d3_df["popularity_score"] = 100 * (
    0.40 * (1 - d3_df["availability_30"] / 30.0)
    + 0.30 * (1 - d3_df["availability_60"] / 60.0)
    + 0.20 * (1 - d3_df["availability_90"] / 90.0)
    + 0.10 * (1 - d3_df["availability_365"] / 365.0)
)

d3_updates = [
    UpdateOne(
        {"_id": row["_id"]},
        {
            "$set": {
                "popularity_score": float(row["popularity_score"]),
                "effective_price_pppn_two_guests_week": float(row["effective_price_pppn"]) if pd.notna(row["effective_price_pppn"]) else None,
            }
        },
    )
    for _, row in d3_df.iterrows()
]

if d3_updates:
    listings_collection.bulk_write(d3_updates, ordered=False)

listings_collection.create_index([("popularity_score", DESCENDING)], name="popularity_score_idx")

listings_collection.find_one(
    {"_id": d3_df.iloc[0]["_id"]},
    {"_id": 1, "property_type": 1, "review_stats.rating": 1, "popularity_score": 1, "effective_price_pppn_two_guests_week": 1},
)
```

# E. Property Features Expert

## E0. Shared assumptions and helper setup

These cells implement **E1**, **E2**, and **E3** on top of the BI-ready collections created above.

**Modeling / analytical assumptions used**

1. **City field**: `address.market` (consistent with D).
2. **Property size category (E1)** — based on `accommodates`:
   - **Small**: 1–3 guests (studio / 1 BR)
   - **Medium**: 4–6 guests (2–3 BR)
   - **Large**: 7+ guests (4+ BR or large common areas)
3. **Effective price per person per night (E1)** — assumes *all* spots filled for one week:

   $$
   \text{pppn} = \frac{7 \cdot price + cleaning\_fee + 7 \cdot extra\_people \cdot \max(0,\, accommodates - guests\_included)}{7 \cdot accommodates}
   $$

   Missing fees → 0; missing `guests_included` → 1.
4. **Professional vs Amateur (E2)**: host manages > 1 listing globally → Professional (same as D).
5. **Booking rate (E2)**: `review_stats.count / months_active`, where `months_active` = months between `first_review` and `last_review` (min 1 month). This normalises by each listing's active period, making comparisons fair between new and old listings.
6. **Comfort categories (E3)**: `room_type` × simplified `property_type`.  
   Five most common combos retained.
7. **Size group (E3)**: `(bedrooms + bathrooms) / accommodates` → terciles: Small / Medium / Large.
8. **E3 scope**: amateur hosts only.

```python
# E2C. Baseline query performance analysis (BEFORE optimisation)

# Helper: explain with executionStats verbosity
def explain_with_stats(collection, pipeline):
    return target_database.command({
        "explain": {
            "aggregate": collection.name,
            "pipeline": pipeline,
            "cursor": {}
        },
        "verbosity": "executionStats"
    })

# Improved metric extractor (recursive search for nested agg explain fields)
def extract_metrics(doc, label):
    return {
        "version": label,
        "execution_time_ms": max(recursive_find_values(doc, "executionTimeMillis") or [None]),
        "n_returned": max(recursive_find_values(doc, "nReturned") or [None]),
        "total_docs_examined": max(recursive_find_values(doc, "totalDocsExamined") or [None]),
        "total_keys_examined": max(recursive_find_values(doc, "totalKeysExamined") or [None]),
        "stages": ", ".join(flatten_stage_names(doc)),
        "used_disk": any(recursive_find_values(doc, "usedDisk")) if recursive_find_values(doc, "usedDisk") else None,
    }

# Pipeline under test (core booking-rate calculation)
e2_calc_pipeline = [
    {"$match": {
        "address.market": {"$in": TARGET_CITIES},
        "host.host_id": {"$ne": None},
        "review_stats.count": {"$ne": None},
        "first_review": {"$ne": None},
        "last_review": {"$ne": None},
    }},
    {"$group": {
        "_id": {"city": "$address.market", "host_id": "$host.host_id"},
        "total_reviews": {"$sum": {"$toDouble": {"$ifNull": ["$review_stats.count", 0]}}},
        "n_listings": {"$sum": 1},
    }},
    {"$addFields": {"host_type": {"$cond": [{"$gt":["$n_listings",1]}, "Professional", "Amateur"]}}},
]

# Temporarily drop ALL indexes that could help the $match (to force COLLSCAN)
indexes_to_restore = [
    ("e2_market_host_reviews_idx",  [("address.market",1),("host.host_id",1),("review_stats.count",1),("first_review",1),("last_review",1)]),
    ("host_id_idx",                 [("host.host_id",1)]),
    ("review_count_idx",            [("review_stats.count",-1)]),
    ("host_city_rating_idx",        [("host.host_id",1),("address.market",1),("review_stats.rating",-1)]),
    ("market_property_room_price_idx", [("address.market",1),("property_type",1),("room_type",1),("price",1)]),
]

dropped = []
for name, _ in indexes_to_restore:
    try:
        listings_collection.drop_index(name)
        dropped.append(name)
    except Exception:
        pass
print(f"Temporarily dropped {len(dropped)} indexes: {dropped}")

# Run baseline explain (should now be COLLSCAN)
exp_before = explain_with_stats(listings_collection, e2_calc_pipeline)
m_before = extract_metrics(exp_before, "Before (no index)")

print("\n=== BASELINE PERFORMANCE ===")
display(pd.DataFrame([m_before]))

print("\n=== ANALYSIS ===")
print("Without any supporting indexes, the $match stage performs a full COLLSCAN —")
print("MongoDB must read every document in the collection to evaluate the five filter")
print("conditions. This leads to high execution time and all 5555 documents examined.")
print("A compound index on the $match fields would let MongoDB skip non-matching")
print("documents entirely (IXSCAN), dramatically reducing work.")
```

## E3. Property Size and Property Comfort

```python
# E3A. Build dataset for amateur hosts — size group + comfort category

e3_pipeline = [
    {"$match": {"host.host_id": {"$ne": None}, "review_stats.rating": {"$ne": None}}},
    {"$project": {
        "_id":1, "host_id":"$host.host_id",
        "accommodates":1, "bedrooms":1, "bathrooms":1,
        "property_type":1, "room_type":1,
        "review_score": {"$toDouble": "$review_stats.rating"},
    }}
]
e3_df = pd.DataFrame(list(listings_collection.aggregate(e3_pipeline)))

for col in ["accommodates","bedrooms","bathrooms"]:
    e3_df[col] = pd.to_numeric(e3_df[col], errors="coerce")
e3_df["accommodates"] = e3_df["accommodates"].fillna(1).clip(lower=1)
e3_df["bedrooms"]     = e3_df["bedrooms"].fillna(0)
e3_df["bathrooms"]    = e3_df["bathrooms"].fillna(1)

# Amateur hosts only
e3_df["global_count"] = e3_df["host_id"].map(host_listing_count_map)
e3_df = e3_df[e3_df["global_count"] == 1].copy()

# Size group: (bedrooms + bathrooms) / accommodates → terciles
e3_df["size_ratio"] = (e3_df["bedrooms"] + e3_df["bathrooms"]) / e3_df["accommodates"]
e3_df["size_group"]  = pd.qcut(e3_df["size_ratio"], q=3, labels=["Small","Medium","Large"],
                               duplicates="drop")

# Persist size_group to DB (efficient batch per group)
for sg in ["Small","Medium","Large"]:
    ids = e3_df[e3_df["size_group"]==sg]["_id"].tolist()
    if ids:
        listings_collection.update_many({"_id":{"$in":ids}}, {"$set":{"size_group":sg}})
print(f"size_group written to {len(e3_df)} amateur listings.")

# Comfort category: room_type x simplified property_type
def simplify_pt(pt):
    p = str(pt).lower()
    if any(k in p for k in ["apartment","aparthotel","condo","loft"]): return "Apartment"
    if any(k in p for k in ["house","villa","townhouse","bungalow"]): return "House"
    if any(k in p for k in ["hotel","boutique","resort"]): return "Hotel"
    if any(k in p for k in ["hostel","dorm"]): return "Hostel"
    return "Other"

def simplify_rt(rt):
    r = str(rt).lower()
    if "entire" in r: return "Entire"
    if "private" in r: return "Private Room"
    return "Shared/Other"

e3_df["comfort_category"] = e3_df["room_type"].apply(simplify_rt) + " " + e3_df["property_type"].apply(simplify_pt)

# Top 5 comfort categories
top5 = e3_df["comfort_category"].value_counts().head(5).index.tolist()
e3_plot = e3_df[e3_df["comfort_category"].isin(top5)].copy()
e3_plot = e3_plot[e3_plot["review_score"].notna() & np.isfinite(e3_plot["review_score"])].copy()

print(f"Amateur listings with valid data: {len(e3_plot)}")
print(f"\nSize groups:\n{e3_plot['size_group'].value_counts().to_string()}")
print(f"\nTop 5 comfort categories:\n{e3_plot['comfort_category'].value_counts().head(5).to_string()}")
```
